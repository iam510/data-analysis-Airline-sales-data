```python
import pandas as pd

# 读取.csv文件
df = pd.read_csv('表1.csv',encoding='gbk')

# 删除多列
columns_to_drop = ['出发站三字码', '到达站三字码', '航司','飞行航班']
df = df.drop(columns_to_drop, axis=1)

# 将日期列转换为datetime类型
df['航班日期'] = pd.to_datetime(df['航班日期'])
df['批次时间（batch_date）'] = pd.to_datetime(df['批次时间（batch_date）'])

df
```

```python
# 获取当前列的顺序
column_order = df.columns.tolist()
column_order
```

```python
# 指定新的列顺序
new_column_order = ['航班号',
 '航班日期',
 '批次时间（batch_date）',
 '实际销售额（sale_volume）',
 '航线',
 '客座率',
 '实际座公里收入',
 '舱位容量',
 '订座数',
 '行距',
 '客运公里',
 '座位公里',
 '机型',
 '起飞时间',
 '到达时间',
 '客公里收入',
 '销售日期',
 'J舱订座数',
 'W舱订座数',
 'Y舱订座数',
 '燃油附加费']

# 重排列顺序
df = df[new_column_order]
df
```

```python
# 按照两列进行排序
sorted_df = df.sort_values(by=['航班号', '航班日期','批次时间（batch_date）'])
sorted_df
```

```python
sorted_df.to_csv('sorted_filename.csv', index=False,encoding='gbk')
```
