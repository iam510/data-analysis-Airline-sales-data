```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'SimSun'

# 读取.csv文件
df = pd.read_csv('table2.csv',encoding='gbk')

# 将日期列转换为日期类型
df['航班日期'] = pd.to_datetime(df['航班日期'])

# 按月份计算销售额总和
monthly_sales = df.groupby(df['航班日期'].dt.to_period('M'))['航线金额'].sum()
# 按周计算销售额总和
weekly_sales = df.resample('W', on='航班日期')['航线金额'].sum()
# 按日期计算销售额总和
daily_sales = df.groupby(df['航班日期'])['航线金额'].sum()

# 将Period类型转换为字符串格式
monthly_sales.index = monthly_sales.index.astype(str)
```

```python
plt.plot(monthly_sales.index, monthly_sales)
plt.xlabel('月份')
plt.ylabel('销售额')
plt.title('每月销售额变化')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(weekly_sales.index, weekly_sales)
plt.xlabel('周')
plt.ylabel('销售额')
plt.title('每周销售额变化')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(daily_sales.index, daily_sales)
plt.xlabel('日期')
plt.ylabel('销售额')
plt.title('每天销售额变化')
plt.xticks(rotation=45)
plt.show()
```
