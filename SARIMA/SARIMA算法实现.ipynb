```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# 读取数据
data = pd.read_csv('../data.csv',encoding='gbk')

# 将日期列转换为日期时间类型
data['航班日期'] = pd.to_datetime(data['航班日期'])

# 将日期列设置为索引
data.set_index('航班日期', inplace=True)

# 检查数据
print(data.head())

# 绘制原始数据
plt.plot(data['航线金额'])
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.title('Flight Sales')
plt.show()

# 拟合SARIMA模型
model = SARIMAX(data['航线金额'], order=(1, 1, 1), seasonal_order=(0, 1, 1, 4))
model_fit = model.fit()

# 预测
forecast = model_fit.predict(start='2024-01-01', end='2024-10-31')

# 绘制预测结果
plt.plot(data['航线金额'], label='Actual')
plt.plot(forecast, label='Forecast')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.title('Flight Sales Forecast')
plt.legend()
plt.show()
```

原数据没有平稳性、是非线性的、并且有一些异常值,并不适合用ARIMA模型去拟合，模型选择错误。
如果数据量增多至多年的销售数据，该模型可能会有比较好的效果。
