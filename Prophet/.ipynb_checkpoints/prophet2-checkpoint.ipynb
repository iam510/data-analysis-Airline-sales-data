```python
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'SimSun'

# 读取数据集
data = pd.read_csv('../data.csv',encoding='gbk')

# 将数据转换为Prophet所需的格式
df = pd.DataFrame()
df['ds'] = pd.to_datetime(data['航班日期'])
df['y'] = data['航线金额']

# 划分训练集和测试集
train_data = df[:292]
test_data = df[292:]

# 创建并训练Prophet模型
model = Prophet()
model.fit(train_data)

# 预测未来的73天
future = model.make_future_dataframe(periods=73)
forecast = model.predict(future)

# 提取预测结果和真实结果
predicted = forecast['yhat'][292:]
actual = test_data['y']

# 绘制折线图
plt.plot(actual, label='Actual')
plt.plot(predicted, label='Predicted')
plt.xlabel('Day')
plt.ylabel('Amount')
plt.title('Prophet Model - Actual vs Predicted')
plt.legend()
plt.show()
```
