```python
import pandas as pd

# 读取.csv文件
df = pd.read_csv('表1.csv',encoding='gbk')
df.columns.tolist()
```

```python
# 删除多列
columns_to_drop = ['航线',
                 '客座率',
                 '实际座公里收入',
                 '出发站三字码',
                 '到达站三字码',
                 '航司',
                 '舱位容量',
                 '订座数',
                 '行距',
                 '客运公里',
                 '座位公里',
                 '机型',
                 '起飞时间',
                 '到达时间',
                 '客公里收入',
                 '销售日期',
                 '飞行航班',
                 'J舱订座数',
                 'W舱订座数',
                 'Y舱订座数',
                 '燃油附加费']
df = df.drop(columns_to_drop, axis=1)
df
```

```python
# 将日期列转换为datetime类型
df['航班日期'] = pd.to_datetime(df['航班日期'])
df['批次时间（batch_date）'] = pd.to_datetime(df['批次时间（batch_date）'])
sorted_df = df.sort_values(by=[ '航班号','航班日期','批次时间（batch_date）'])
sorted_df = sorted_df.reset_index(drop=True)
sorted_df
```

```python
rows_to_keep = []

# 遍历数据中的每一行
for index, row in sorted_df.iterrows():
    if index < len(sorted_df) - 1:  # 检查是否为最后一行
        if ((row['航班号'] != sorted_df.iloc[index+1]['航班号']) or (row['航班日期'] != sorted_df.iloc[index+1]['航班日期'])):
            # 如果当前行与下一行的航班号或航班日期不相同，则将当前行的索引添加到列表中
            rows_to_keep.append(index)
    else:
        # 最后一行的索引添加到列表中
        rows_to_keep.append(index)

# 通过索引筛选出符合要求的行，构建新的 DataFrame
result_df = sorted_df.loc[rows_to_keep].reset_index(drop=True)

# 输出处理后的数据
result_df
```

```python
last_df = result_df.sort_values(by=['航班日期','航班号'])
last_df = last_df.reset_index(drop=True)
last_df
```

```python
last_df = last_df.drop(['批次时间（batch_date）'], axis=1)
new_column_order = [
 '航班日期',
 '航班号',
 '实际销售额（sale_volume）'
 ]

# 重排列顺序
last_df = last_df[new_column_order]
last_df
```

```python
last_df.to_csv('table1.csv', index=False,encoding='gbk')
```
