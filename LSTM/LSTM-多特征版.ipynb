```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
plt.rcParams['font.family'] = 'SimSun'

# 读取数据集
data = pd.read_csv('../newData.csv',encoding='gbk')
# 转换日期特征为数值型
data['航班日期'] = pd.to_datetime(data['航班日期'])
data['航班日期'] = (data['航班日期'] - data['航班日期'].min()).dt.days
```

```python
# 选择特征和目标变量
features = data[['航班日期', '月份', '日期', '星期几', '季节', '节假日']].values
target = data['航线金额'].values

# 归一化特征和目标变量
scaler = MinMaxScaler(feature_range=(0, 1))
features_scaled = scaler.fit_transform(features)
target_scaled = scaler.fit_transform(target.reshape(-1, 1))

# 调整输入数据的形状
time_steps = 10  # 设置时间步数
input_features = []
input_target = []
for i in range(len(features_scaled) - time_steps):
    input_features.append(features_scaled[i:i+time_steps])
    input_target.append(target_scaled[i+time_steps])
input_features = np.array(input_features)
input_target = np.array(input_target)

# 划分训练集和测试集
train_size = int(len(input_features) * 0.8)
train_features = input_features[:train_size]
train_target = input_target[:train_size]
test_features = input_features[train_size:]
test_target = input_target[train_size:]

# 调整模型的输入形状
model = Sequential()
model.add(LSTM(units=50, input_shape=(train_features.shape[1], train_features.shape[2])))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# 训练模型
model.fit(train_features, train_target, epochs=50, batch_size=32)

# 在测试集上评估模型
predictions = model.predict(test_features)
predictions_scaled = scaler.inverse_transform(predictions)
test_target_unscaled = scaler.inverse_transform(test_target)

# 反归一化预测结果
predictions_unscaled = predictions_scaled.reshape(-1)
```

```python
# 绘制折线图
plt.plot(predictions_unscaled, label='预测')
plt.plot(test_target_unscaled, label='实际')
plt.xlabel('样本索引')
plt.ylabel('航线金额')
plt.title('航线金额预测结果')
plt.legend()
plt.show()
```
