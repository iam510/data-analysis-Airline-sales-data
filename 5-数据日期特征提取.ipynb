```python
import pandas as pd
import holidays

# 加载原始数据
data = pd.read_csv('data.csv',encoding='gbk')

# 转换日期列为日期时间类型
data['航班日期'] = pd.to_datetime(data['航班日期'])

# 提取航班日期特征
data['月份'] = data['航班日期'].dt.month
data['日期'] = data['航班日期'].dt.day
data['星期几'] = data['航班日期'].dt.dayofweek
data['季节'] = data['航班日期'].dt.quarter

# 获取节假日信息
holiday_list = holidays.China(years=data['航班日期'].dt.year.unique())

# 添加是否为节假日的列
data['节假日'] = data['航班日期'].dt.date.astype(str).map(lambda x: x in holiday_list)

# 创建新的数据集
new_data = data[['航班日期','航线金额', '月份', '日期', '星期几', '季节', '节假日']]
new_data
```

```python
# 可以根据需要保存新的数据集到文件
new_data.to_csv('newData.csv', index=False,encoding='gbk')
```
