```python
import pandas as pd

# 读取.csv文件
df = pd.read_csv('表2.csv',encoding='gbk')
df.columns.tolist()
```

```python
# 删除多列
columns_to_drop = ['订票日期',
 '订票时间',
 '距离起飞天数',
 '起飞机场三字码',
 '中转机场三字码',
 '到达机场三字码',
 '主舱位代码',
 '子舱位代码',
 '支付金额',
 '票号',
 '主键，票号加票联号',
 '年份字段']
df = df.drop(columns_to_drop, axis=1)
df = df.drop(df.index[0])
df
```

```python
df['航班号'] = df['航班号'].astype(str)
df['航班号']=df['承运人']+df['航班号']
df = df.drop(['承运人'], axis=1)
df
```

```python
df['航班日期'] = pd.to_datetime(df['航班日期'])
sorted_df = df.sort_values(by=[ '航班日期','航班号'])
sorted_df = sorted_df.reset_index(drop=True)
sorted_df
```

```python
sorted_df['航线金额'] = sorted_df['航线金额'].astype(float)
grouped_df = sorted_df.groupby([ '航班日期','航班号'])['航线金额'].sum().reset_index()
grouped_df
```

```python
grouped_df.to_csv('table2.csv', index=False,encoding='gbk')
```
