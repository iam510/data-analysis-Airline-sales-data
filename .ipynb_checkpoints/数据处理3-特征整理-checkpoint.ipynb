```python
import pandas as pd

# 读取.csv文件
df = pd.read_csv('airlineSales.csv',encoding='gbk')

df.columns.tolist()
```

```python
df['航班日期'] = pd.to_datetime(df['航班日期'])
sorted_df = df.sort_values(by=[ '航班日期'])
sorted_df
```

```python
columns_to_drop = ['航班号','客座率','实际座公里收入','舱位容量','订座数','客公里收入']
sorted_df = sorted_df.drop(columns_to_drop, axis=1)
sorted_df
```

```python
sorted_df.to_csv('airlineSalesS.csv', index=False,encoding='gbk')
```
