```python
import pandas as pd

# 读取.csv文件
df = pd.read_csv('sorted_filename.csv',encoding='gbk')

df.columns.tolist()
```

```python
# 删除多列
columns_to_drop = ['航线',
                   '行距',
                   '客运公里',
                   '座位公里',
                   '机型',
                   '起飞时间',
                   '到达时间',
                   '销售日期',
                   'J舱订座数',
                   'W舱订座数',
                   'Y舱订座数',
                   '燃油附加费']
df = df.drop(columns_to_drop, axis=1)
df
```

```python
#保留信息最完整的批次行，删除其他多余行
# 创建一个空的列表，用于存储符合要求的行的索引
rows_to_keep = []

# 遍历数据中的每一行
for index, row in df.iterrows():
    if index < len(df) - 1:  # 检查是否为最后一行
        if ((row['航班号'] != df.iloc[index+1]['航班号']) or (row['航班日期'] != df.iloc[index+1]['航班日期'])):
            # 如果当前行与下一行的航班号或航班日期不相同，则将当前行的索引添加到列表中
            rows_to_keep.append(index)
    else:
        # 最后一行的索引添加到列表中
        rows_to_keep.append(index)

# 通过索引筛选出符合要求的行，构建新的 DataFrame
result_df = df.loc[rows_to_keep].reset_index(drop=True)

# 输出处理后的数据
print(result_df)
```

```python
result_df = result_df.drop(['批次时间（batch_date）'], axis=1)
```

```python
result_df.to_csv('airlineSales.csv', index=False,encoding='gbk')
```
